```python
from scipy.io import loadmat
import numpy as np
import json
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import scale
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score
from operator import itemgetter
#%matplotlib notebook

Problem1Data = loadmat("dataset.mat")
# print(json.dumps(Problem1Data, indent=3))
X_Data = Problem1Data['X'].tolist() # Reading X Data
Y_Data1 = Problem1Data['Y'].tolist() # Reading Y Data
Y_Data = []
for values in Y_Data1:
    Y_Data.append(values[0])
#print(Y_Data)
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X_Data, Y_Data, test_size = 0.5, random_state = 9) # Test Train Split
```

```python
# Calculating Linear SVM

C = [1E-10,1E-9,1E-8,1E-7,1E-6,1E-5,1E-4,1E-3,1E-2,1E-1,1.0,2.0,3.0,4.0]
Values = []

for C_Entries in C:
    for RandomState in range(0,31):
        Dir = {}
        Dir['C'] = C_Entries
        Dir['Random_State'] = RandomState
        sc = StandardScaler()
        X_Train = sc.fit_transform(X_Train)
        X_Test = sc.transform(X_Test)
        classifier = SVC(kernel = 'linear', random_state = RandomState, C=C_Entries) #Declaring SVM Kernel
        classifier.fit(X_Train, Y_Train) # Training
        Y_Pred = classifier.predict(X_Test) # Prediction
        Dir['Accuracy'] = accuracy_score(Y_Test, Y_Pred)
        Values.append(Dir)

df = pd.DataFrame(Values)

threedee = plt.figure().add_subplot(projection='3d')
threedee.scatter3D(df['C'], df['Random_State'], df['Accuracy'], color='r')
threedee.set_xlabel('C')
threedee.set_ylabel('Random_State')
threedee.set_zlabel('Accuracy')
plt.title('Linear')
plt.show()

Values = sorted(Values, key=itemgetter('Accuracy'), reverse=True)
print("Hyper-parameters for highest accuracy for Linear SVM")
print(json.dumps(Values[0], indent=3))
```

```python
# Calculate Polynomial SVM Kernel

C = [1E-10,1E-9,1E-8,1E-7,1E-6,1E-5,1E-4,1E-3,1E-2,1E-1,1.0,2.0,3.0,4.0]
Dimension = [1,2,3,4,5]
Values = []

for C_Entries in C:
    for Degree in Dimension:
        Dir = {}
        Dir['C'] = C_Entries
        Dir['Degree of Polynomial'] = Degree
        sc = StandardScaler()
        X_Train = sc.fit_transform(X_Train)
        X_Test = sc.transform(X_Test)
        classifier = SVC(kernel = 'poly', random_state = 0, degree=Degree, C=C_Entries) # Declaring Polynomial SVM Kernel
        classifier.fit(X_Train, Y_Train) # Training
        Y_Pred = classifier.predict(X_Test) # Prediction
        Dir['Accuracy'] = accuracy_score(Y_Test,Y_Pred)
        Values.append(Dir)

df = pd.DataFrame(Values)

threedee = plt.figure().add_subplot(projection='3d')
threedee.scatter3D(df['C'], df['Degree of Polynomial'], df['Accuracy'], color='r')
threedee.set_xlabel('C')
threedee.set_ylabel('Degree of Polynomial')
threedee.set_zlabel('Accuracy')
plt.title('Polynomial')
plt.show()

Values = sorted(Values, key=itemgetter('Accuracy'), reverse=True)
print("Hyper-parameters for highest accuracy for Polynomial SVM")
print(json.dumps(Values[0], indent=3))
```
