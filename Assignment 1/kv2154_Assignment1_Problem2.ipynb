```python
#Solution to Problem 2

from scipy.io import loadmat
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import json
from operator import itemgetter
import gc

Problem2Data = loadmat(r"problem2.mat")
X_Data = Problem2Data['x']
Y_Data = Problem2Data['y']
#print(X_Data)

# for values in range(len(X_Data)):
#     X_Data[values].insert(0,1)

#X_Data = np.array(X_Data)
```

```python
def Convert_Vector(X, Y, Lambda):
    X_Vector = X.tolist()
    for values in range(len(X_Vector)):
        X_Vector[values].insert(0,1)

    X_Vector = np.array(X_Vector)
    XTX = np.dot(X_Vector.T, X_Vector)
    XTXLambda = XTX + (Lambda * np.identity(len(XTX)))
    Inverse = np.linalg.pinv(XTXLambda)
    Theta = np.dot(np.dot(Inverse, X_Vector.T), Y)
    Normalized_Theta = np.linalg.norm(Theta)

    return X_Vector, Normalized_Theta, Theta
```

```python
OptimalLambdaList = []
for i in range(1, 31):
    X_Train, X_Test, Y_Train, Y_Test = train_test_split(X_Data, Y_Data, test_size=0.8, random_state=i)
    ValueList = []

    for Lambda in range(0, 1001):
        ValueDir = {}
        ValueDir['Lambda'] = Lambda

        X_Vector_Train, Normalized_Theta1, Theta1 = Convert_Vector(X_Train, Y_Train, Lambda)
        Hypothesis1 = np.dot(X_Vector_Train, Theta1)
        TrainLoss = (0.5 / X_Train.shape[0]) * (np.sum(np.power(Y_Train - Hypothesis1, 2)) + (Lambda * Normalized_Theta1 * Normalized_Theta1))
        ValueDir['TrainLoss'] = TrainLoss
        ValueDir['TrainLossLog'] = math.log(TrainLoss)

        X_Vector_Test, Normalized_Theta2, Theta2 = Convert_Vector(X_Test, Y_Test, Lambda)
        Hypothesis2 = np.dot(X_Vector_Test, Theta1)
        TestLoss = (0.5 / X_Test.shape[0]) * (np.sum(np.power(Y_Test - Hypothesis2, 2)) + (Lambda * Normalized_Theta1 * Normalized_Theta1))
        ValueDir['TestLoss'] = TestLoss
        ValueDir['TestLossLog'] = math.log(TestLoss)

        ValueDir['DifferenceOfLoss'] = ValueDir['TrainLoss'] - ValueDir['TestLoss']
        ValueDir['DifferenceOfLossLog'] = math.log(abs(ValueDir['TestLoss'] - ValueDir['TrainLoss']))
        ValueList.append(ValueDir)

    #print(json.dumps(ValueList, indent=3))
    MinLossSort = sorted(ValueList, key=itemgetter('TestLoss'))
    LambdaOptimal = MinLossSort[0]['Lambda']
    print(json.dumps(MinLossSort[0], indent=3))
    print(LambdaOptimal)
    OptimalLambdaList.append(LambdaOptimal)

    DataFrame = pd.DataFrame(ValueList)
    DataFrame.plot('Lambda', ['TrainLoss', 'TestLoss'])
    plt.plot(LambdaOptimal, ValueList[LambdaOptimal]['TestLoss'], 'xb')
    plt.text(LambdaOptimal, ValueList[LambdaOptimal]['TestLoss'], "(Lambda={}, {})".format(str(LambdaOptimal), str(MinLossSort[0]['TestLoss'])))
    plt.title('Cross-validation at Random State = {}'.format(i))
    plt.show()
    gc.collect()

# print(OptimalLambdaList)
# print("Optimal Lambda = {}".format(int(sum(OptimalLambdaList) / len(OptimalLambdaList))))
```
